```python
big_box = cv2.imread('../datasets/train/images/Subject_0_0.png', cv2.IMREAD_GRAYSCALE)
canvas_height, canvas_wdith = 1000, 1000
canvas = np.zeros((canvas_height, canvas_wdith), dtype=np.uint8)

small_box_xcenter, small_box_ycenter, small_box_width, small_box_height = 500, 500, 500, 200
small_box = cv2.resize(big_box, (small_box_width, small_box_height))

xmin = small_box_xcenter - small_box_width // 2
ymin = small_box_ycenter - small_box_height // 2

canvas[ymin:ymin+small_box_height, xmin:xmin+small_box_width] = small_box
plt.imshow(canvas)
```
